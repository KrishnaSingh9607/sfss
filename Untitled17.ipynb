```python

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_colwidth', None)
```

```python

df = pd.read_csv(r"C:\Users\kunal\Downloads\Global_Pollution_Analysis.csv")
```

```python


df.columns = df.columns.str.strip()

```

```python

print("Columns:", df.columns.tolist())
print(df.head(3))
```

```python

num_cols = [
    'Air_Pollution_Index','Water_Pollution_Index','Soil_Pollution_Index',
    'Industrial_Waste (in tons)','Energy_Recovered (in GWh)','CO2_Emissions (in MT)',
    'Renewable_Energy (%)','Plastic_Waste_Produced (in tons)',
    'Energy_Consumption_Per_Capita (in MWh)','Population (in millions)','GDP_Per_Capita (in USD)'
]
for col in num_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')


for col in num_cols:
    if col in df.columns:
        df[col].fillna(df[col].median(), inplace=True)

```

```python

cat_cols = ['Country','Year']
for col in cat_cols:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip()
        df[col].fillna(df[col].mode().iloc if not df[col].mode().empty else 'Unknown', inplace=True)
```

```python
scaler = MinMaxScaler()
pollution_cols = ['Air_Pollution_Index','Water_Pollution_Index','Soil_Pollution_Index']
existing = [c for c in pollution_cols if c in df.columns]
df[existing] = scaler.fit_transform(df[existing])
```

```python
le_country, le_year = LabelEncoder(), LabelEncoder()
if 'Country' in df.columns:
    df['Country_Encoded'] = le_country.fit_transform(df['Country'].astype(str))
if 'Year' in df.columns:
    df['Year_Encoded'] = le_year.fit_transform(df['Year'].astype(str))
```

```python

ec_col = 'Energy_Consumption_Per_Capita (in MWh)'
pop_col = 'Population (in millions)'
gdp_col = 'GDP_Per_Capita (in USD)'

if ec_col in df.columns and gdp_col in df.columns:
    df['Energy_per_GDP'] = df[ec_col] / df[gdp_col]
if ec_col in df.columns and pop_col in df.columns:
    
    df['Energy_Intensity_perM'] = df[ec_col] * df[pop_col]
```

```python


def categorize_pollution(v):
    if pd.isna(v): return 'Unknown'
    if v < 0.33: return 'Low'
    elif v < 0.66: return 'Medium'
    else: return 'High'

for base_col, sev_col in [
    ('Air_Pollution_Index','Air_Severity'),
    ('Water_Pollution_Index','Water_Severity'),
    ('Soil_Pollution_Index','Soil_Severity')
]:
    if base_col in df.columns:
        df[sev_col] = df[base_col].apply(categorize_pollution)
```

```python


er_col = 'Energy_Recovered (in GWh)'
if er_col in df.columns:
    ser = pd.to_numeric(df[er_col], errors='coerce').fillna(df[er_col].median())
    df['Energy_Recovery_Category'] = pd.qcut(ser, q=3, labels=['Low','Medium','High'], duplicates='drop')
```

```python

trend = df.groupby(['Country','Year'], as_index=False)[['Air_Pollution_Index','Water_Pollution_Index','Soil_Pollution_Index']].mean()
trend.head()
```

```python

required = ['Air_Severity','Water_Severity','Energy_Recovery_Category']
missing = [c for c in required if c not in df.columns]
print("Missing for transactions:", missing)
```

```python


def txn_row(row):
    items = []
    items.append('Air_' + (str(row['Air_Severity']) if pd.notna(row.get('Air_Severity')) else 'Unknown'))
    items.append('Water_' + (str(row['Water_Severity']) if pd.notna(row.get('Water_Severity')) else 'Unknown'))
    if 'Soil_Severity' in row.index:
        items.append('Soil_' + (str(row['Soil_Severity']) if pd.notna(row.get('Soil_Severity')) else 'Unknown'))
    if 'Energy_Recovery_Category' in row.index:
        items.append('EnergyRec_' + (str(row['Energy_Recovery_Category']) if pd.notna(row.get('Energy_Recovery_Category')) else 'Unknown'))
    
    if 'Renewable_Energy (%)' in row.index:
        items.append('Renewables_' + ('High' if row['Renewable_Energy (%)']>=0.66 else ('Medium' if row['Renewable_Energy (%)']>=0.33 else 'Low')))
    if 'CO2_Emissions (in MT)' in row.index:
        
        pass
    return items

transactions = df.apply(txn_row, axis=1).tolist()

```

```python

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_apriori = pd.DataFrame(te_ary, columns=te.columns_)
print(df_apriori.shape, df_apriori.head(3))
```

```python

frequent_itemsets = apriori(df_apriori, min_support=0.05, use_colnames=True)
frequent_itemsets = frequent_itemsets.sort_values('support', ascending=False)
display(frequent_itemsets.head(10))
```

```python


rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.6)
rules = rules.sort_values(['lift','confidence','support'], ascending=False)
display(rules.head(10))
```

```python

strong_rules = rules[(rules['lift'] >= 1.2) & (rules['confidence'] >= 0.7)]
display(strong_rules[['antecedents','consequents','support','confidence','lift']].head(15))
```

```python


df_apriori_a = df_apriori.iloc[::2]
df_apriori_b = df_apriori.iloc[1::2]

fi_a = apriori(df_apriori_a, min_support=0.05, use_colnames=True)
fi_b = apriori(df_apriori_b, min_support=0.05, use_colnames=True)

rules_a = association_rules(fi_a, metric='confidence', min_threshold=0.6)
rules_b = association_rules(fi_b, metric='confidence', min_threshold=0.6)

```

```python

pairs_a = set(zip(rules_a['antecedents'].astype(str), rules_a['consequents'].astype(str)))
pairs_b = set(zip(rules_b['antecedents'].astype(str), rules_b['consequents'].astype(str)))
overlap = pairs_a & pairs_b
print("Rule pair overlap (holdout):", len(overlap))
```

```python

plt.figure(figsize=(7,4))
sns.histplot(frequent_itemsets['support'], bins=20, kde=False)
plt.title('Frequent Itemsets Support Distribution')
plt.xlabel('Support'); plt.ylabel('Count')
plt.show()
```

```python


plt.figure(figsize=(7,5))
sns.scatterplot(data=rules, x='support', y='confidence', size='lift', sizes=(20,200), alpha=0.6)
plt.title('Association Rules: Support vs Confidence (size=Lift)')
plt.show()


```

```python

topn = rules.nlargest(15, 'lift')
plt.figure(figsize=(8,6))
sns.barplot(y=topn.apply(lambda r: f"{list(r['antecedents'])}→{list(r['consequents'])}", axis=1),
            x=topn['lift'], orient='h')
plt.title('Top 15 Rules by Lift'); plt.xlabel('Lift'); plt.ylabel('Rule')
plt.tight_layout(); plt.show()
```
